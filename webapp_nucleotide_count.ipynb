```python
# input text box
st.header('Enter DNA Sequence')

sequence_input = ">NG_005130.1 Rattus norvegicus pseudogene for small nuclear RNA U3 (H3.9) (LOC366310) on chromosome 5 TCCGTGTAAAATAAGAGTGGAATCACCTCACAGATGAAAAGTACAGGTGAAAGACTATACTTTCAGGGAT CATTTCTATAGTTCGTTACTAGAGAAGTTTCTCTGACTGTGTAGAGCACCCGAAA"
```

```python
# sequence = st.sidebar.txt_area("Sequence input", sequence_input, 
sequence = st.text_area("Sequence input", sequence_input, height =25)
```

```python
sequence = sequence.splitlines()
sequence
```

```python
sequence = sequence[1:] # skip the sequence name (first line)
sequence
```

```python
sequence = ' '.join(sequence) # concatenates list to string
```

```python
st.write("""
***
""")
```

```python
# print the input dna sequence
st.header('INPUT (DNA Quary)')
sequence
```

```python
# DNA Nucleotide count
st.header('OUTPUT (DNA Nucleotide Count)')
```

```python
# print dictionary
st.subheader('Print Dictionary')
```

```python
def DNA_Nucleotide_Count(seq):
    d = dict([
        ('A', seq.count('A')),
        ('T', seq.count('T')),
        ('G', seq.count('G')),
        ('C', seq.count('C'))
    ])

    return d
```

```python
X = DNA_Nucleotide_Count(sequence)
```

```python
X_lable = list(X)
```

```python
X_values = list(X.values())
```

```python
X
```

```python
# print txt
st.subheader('2. Print txt')
```

```python
st.write('There are ' + str(X['A']) + ' Adenine (A)')
st.write('There are ' + str(X['T']) + ' Thymine (T)')
st.write('There are ' + str(X['G']) + ' Adenine [Guanine (G)]')
st.write('There are ' + str(X['C']) + ' Thymine [Cytosine (C)]')
```

```python
# display dataframe
st.subheader('3. Display DataFrame')
```

```python
df = pd.DataFrame.from_dict(X, orient = 'index')
```

```python
df = df.rename({0: 'count'}, axis = 'columns')
```

```python
df.reset_index(inplace = True)
```

```python
df = df.rename(columns = {'index' : 'Nucleotide'})
```

```python
st.write(df)
```

```python
# display bar chart useing alter
st.subheader('4. Display Bar Chart ')
```

```python
p = alt.Chart(df).mark_bar().encode(
    x = 'Nucleotide',
    y = 'count'
)
```

```python
p = p.properties(
    width=alt.Step(80) # control width of bar
)
```
